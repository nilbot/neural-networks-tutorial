```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline
import numpy as np

# source file mlp.py
import mlp
```

```python
custom_data_home = "data/"
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original', data_home=custom_data_home)
target = mlp.to_indexmatrix(10, mnist.target)
mnist.target = target
```

```python
mnist = mlp.get_MNIST()
```

```python
training_X, training_y, testing_X, testing_y = mlp.get_data_split(mnist.data, mnist.target, 0.8)
```

```python
print(training_X.shape)
print(testing_X.shape)
```

```python
naive500 = mlp.MLP(epsilon=0.01,
                   X=training_X,
                   y=training_y,
                   hu_size=300,
                  )
```

```python
naive500.train(batch_size=16, testset_X=testing_X, testset_y=testing_y, print_loss=True)
```

```python
naive300 = mlp.MLP(epsilon=0.0002,
                   X=training_X,
                   y=training_y,
                   hu_size=300,
                  )
acc_log, loss = naive300.train(batch_size=16, testset_X=testing_X, testset_y=testing_y, print_loss=True)
```

```python
plt.plot([i[0] for i in acc_log], [i[1] for i in acc_log])
plt.plot([i[0] for i in acc_log], np.zeros((len(acc_log), 1))+0.80)
plt.plot([i[0] for i in acc_log], np.zeros((len(acc_log), 1))+0.85)
plt.plot([i[0] for i in acc_log], np.zeros((len(acc_log), 1))+0.90)
plt.plot([i[0] for i in acc_log], np.zeros((len(acc_log), 1))+0.95)
plt.title("Accuracy ~ Epoch HU size {0}, Batch size {1}".format(300,16))
```
